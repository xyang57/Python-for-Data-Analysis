## Figures and Subplots

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(),'k--')
_ = ax1.hist(np.random.randn(100), bins = 20, color = 'k', alpha = 0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
```

```python
fig, axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins = 50, color = 'k', alpha = 0.5)
plt.subplots_adjust(wspace = 0, hspace = 0)
```

## Colors, Markers, and Line Styles

```python
plt.plot(np.random.randn(30).cumsum(),'ko--')
```

```python
data = np.random.randn(30).cumsum()
```

```python
plt.plot(data, color = 'k', linestyle = 'dashed', marker = 'o', label = 'Default')
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

## Ticks, Labels, and Legends

### Setting the title, axis labels, ticks and ticklabels

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])
```

```python
labels = ax.set_xticklabels(['one','two','three','four','five'], rotation = 30, fontsize = 'small')
```

```python
ax.set_title('My first matplotlib plot')
```

```python
ax.set_xlabel('Stages')
```

```python
fig
```

```python
props = {
    'ylabel': 'Y axis'
}
```

```python
ax.set(**props)
```

```python
fig
```

### Adding legends

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
ax.plot(np.random.randn(1000).cumsum(), 'k', label = 'one')
```

```python
ax.plot(np.random.randn(1000).cumsum(), 'k--', label = 'two')
```

```python
ax.plot(np.random.randn(1000).cumsum(), 'k.', label = 'three')
```

```python
ax.legend(loc = 'best')
```

```python
fig
```

```python
fig = plt.figure(figsize=(12, 6)); ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.savefig('figpath.png', dpi=400, bbox_inches='tight')
```

## Scatter or Point Plots

```python
marco = pd.read_csv('examples/macrodata.csv')
```

```python
data = marco[['cpi','m1','tbilrate','unemp']]
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data[-5:]
```

```python
trans_data.diff?
```

```python
sns.regplot('m1','unemp', data = trans_data)
plt.title('Changes in log m1 versus log unemp')
```

```python
sns.pairplot(trans_data, diag_kind = 'kde', plot_kws = {'alpha':0.2})
```

## Facet Grids and Categorical Data

```python
sns.factorplot(x = 'day', y = 'tip_pct', hue = 'time', col = 'smoker',
              kind = 'bar', data = tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x = 'day', y = 'tip_pct', row = 'time', col = 'smoker',
          kind = 'bar', data = tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x = 'tip_pct', y = 'day', kind = 'box', data = tips[tips.tip_pct < 0.3])
```
